```python
import numpy as np

# 1. Pull Reliance and 'Market' (Nifty) data from your Oracle DB
query = """
SELECT TICKER, PRICE_DATE, CLOSE_PRICE 
FROM MARKETDATA 
WHERE TICKER IN ('RELIANCE.NS', 'NIFTY50_INDEX') 
ORDER BY PRICE_DATE
"""
df_risk = pd.read_sql(query, engine)
df_risk.columns = [c.upper() for c in df_risk.columns]

# 2. Calculate Daily Returns (CFA Concept)
df_pivot = df_risk.pivot(index='PRICE_DATE', columns='TICKER', values='CLOSE_PRICE')
returns = df_pivot.pct_change().dropna()

# 3. Calculate BETA (CFA: Systematic Risk)
covariance = returns.cov().iloc[0, 1]
market_variance = returns['NIFTY50_INDEX'].var()
beta = covariance / market_variance

# 4. Calculate SHARPE RATIO (CFA: Risk-Adjusted Return)
risk_free_rate = 0.07 / 252 # Assuming 7% annual ROI (Standard for India)
excess_return = returns['RELIANCE.NS'].mean() - risk_free_rate
sharpe_ratio = (excess_return / returns['RELIANCE.NS'].std()) * np.sqrt(252)

print(f"--- GS Project: Risk Analysis ---")
print(f"Reliance Beta: {beta:.2f} (Market = 1.0)")
print(f"Reliance Sharpe Ratio: {sharpe_ratio:.2f}")
```

```python
import numpy as np

# 1. Pull ALL stocks + Nifty Index
query = "SELECT TICKER, PRICE_DATE, CLOSE_PRICE FROM MARKETDATA ORDER BY PRICE_DATE"
df_all = pd.read_sql(query, engine)
df_all.columns = [c.upper() for c in df_all.columns]

# 2. Pivot and Calculate Returns
df_pivot = df_all.pivot(index='PRICE_DATE', columns='TICKER', values='CLOSE_PRICE')
returns = df_pivot.pct_change().dropna()

# 3. Bulk Risk Calculations (CFA: Portfolio Management)
risk_free_rate = 0.07 / 252
risk_metrics = []

for ticker in returns.columns:
    if ticker == 'NIFTY50_INDEX': continue
    
    # Beta Calculation
    cov = returns[[ticker, 'NIFTY50_INDEX']].cov().iloc[0,1]
    mkt_var = returns['NIFTY50_INDEX'].var()
    beta = cov / mkt_var
    
    # Sharpe Ratio Calculation
    excess_ret = returns[ticker].mean() - risk_free_rate
    sharpe = (excess_ret / returns[ticker].std()) * np.sqrt(252)
    
    risk_metrics.append({'TICKER': ticker, 'BETA': beta, 'SHARPE_RATIO': sharpe})

# 4. Show the "Goldman Analyst" Leaderboard
df_metrics = pd.DataFrame(risk_metrics)
print("--- Top 5 Stocks by Sharpe Ratio (Risk-Adjusted Return) ---")
print(df_metrics.sort_values('SHARPE_RATIO', ascending=False).head())
```

```python
# Assuming df_metrics is the DataFrame from your previous step
try:
    # Ensure column names are uppercase to match Oracle
    df_metrics.columns = [c.upper() for c in df_metrics.columns]
    
    # Push to Oracle
    df_metrics.to_sql('RISKMETRICS', engine, if_exists='append', index=False)
    print(f"Successfully uploaded Risk Metrics for {len(df_metrics)} stocks.")
except Exception as e:
    print(f"Upload Error: {e}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Pick the best stock from your RiskMetrics leaderboard
top_stock = df_metrics.sort_values('SHARPE_RATIO', ascending=False).iloc[0]['TICKER']
print(f"Simulating for Top Stock: {top_stock}")

# 2. Get last 1 year of daily returns for this stock
query = f"SELECT CLOSE_PRICE FROM MARKETDATA WHERE TICKER = '{top_stock}' ORDER BY PRICE_DATE"

try:
    df_prices = pd.read_sql(query, engine)
    
    # CRITICAL: Force column names to UPPERCASE to stop the KeyError
    df_prices.columns = [c.upper() for c in df_prices.columns]
    
    # Now we can safely access 'CLOSE_PRICE'
    prices = df_prices['CLOSE_PRICE']
    daily_returns = prices.pct_change().dropna()
    print(f"Success: Pulled {len(prices)} price points for {top_stock}")

except Exception as e:
    print(f"Error fetching prices: {e}")
    if 'df_prices' in locals():
        print("Columns found in database:", df_prices.columns.tolist())

# 3. Simulation Parameters (CFA Concept: Geometric Brownian Motion)
mu = daily_returns.mean()
sigma = daily_returns.std()
last_price = prices.iloc[-1]
num_simulations = 10000
num_days = 30 # Predicting the next month

# 4. Run the 10,000 Simulations
simulation_results = np.zeros((num_days, num_simulations))

for i in range(num_simulations):
    # Generating a random path using Drift and Volatility
    prices_path = [last_price]
    for d in range(num_days - 1):
        next_price = prices_path[-1] * (1 + np.random.normal(mu, sigma))
        prices_path.append(next_price)
    simulation_results[:, i] = prices_path

# 5. Calculate Value at Risk (VaR) at 95% Confidence (CFA Level 1 Topic)
ending_prices = simulation_results[-1, :]
var_95 = np.percentile(ending_prices, 5)
potential_loss = ((last_price - var_95) / last_price) * 100

# 6. Plot the "Fan Chart"
plt.figure(figsize=(12, 6))
plt.plot(simulation_results, color='grey', alpha=0.02) # All 10,000 paths
plt.axhline(y=last_price, color='blue', linestyle='--', label='Current Price')
plt.axhline(y=var_95, color='red', linestyle='--', label=f'95% VaR: {var_95:.2f}')
plt.title(f'Goldman Sachs Project: Monte Carlo Simulation for {top_stock}', fontsize=14)
plt.xlabel('Days into Future')
plt.ylabel('Predicted Price')
plt.legend()
plt.show()

print(f"--- Risk Report for {top_stock} ---")
print(f"Current Price: {last_price:.2f}")
print(f"Predicted 95% VaR (Worst Case): {var_95:.2f}")
print(f"Maximum Potential 30-day Loss: {potential_loss:.2f}%")
```

```python
import gradio as gr
import pandas as pd
import matplotlib.pyplot as plt

# 1. Re-define the list for the dropdown
nifty50_list = [
    "RELIANCE.NS", "TCS.NS", "HDFCBANK.NS", "ICICIBANK.NS", "INFY.NS", 
    "BHARTIARTL.NS", "SBI.NS", "LICI.NS", "ITC.NS", "HINDUNILVR.NS"
]

# 2. Define the Analytics Logic (Talking to Oracle 21c)
def analyze_stock(ticker):
    try:
        # Pull data from your Oracle DB
        query = f"SELECT PRICE_DATE, CLOSE_PRICE FROM MARKETDATA WHERE TICKER = '{ticker}' ORDER BY PRICE_DATE"
        df_ui = pd.read_sql(query, engine)
        df_ui.columns = [c.upper() for c in df_ui.columns]
        
        # Calculate Moving Average
        df_ui['SMA20'] = df_ui['CLOSE_PRICE'].rolling(window=20).mean()
        
        # Create Plot
        fig, ax = plt.subplots(figsize=(10, 4))
        ax.plot(df_ui['PRICE_DATE'], df_ui['CLOSE_PRICE'], label='Market Price', color='blue')
        ax.plot(df_ui['PRICE_DATE'], df_ui['SMA20'], label='Predictive Trend', color='orange')
        ax.set_title(f"Live Predictive Analysis: {ticker}")
        ax.legend()
        return fig
    except Exception as e:
        return f"Error: {e}. Ensure Oracle is connected."

# 3. Create the 'demo' object AND launch it in the same cell
demo = gr.Interface(
    fn=analyze_stock, 
    inputs=gr.Dropdown(choices=nifty50_list, label="Select NIFTY Stock"), 
    outputs="plot",
    title="GS 2027: Predictive Analytics Dashboard",
    description="Interface powered by Oracle 21c Database"
)

demo.launch(inline=True)
```

```python
import gradio as gr
import numpy as np

def master_analytics(ticker):
    # 1. Pre-initialize outputs to avoid 'NameError'
    fig_price, ax = plt.subplots(figsize=(10, 4))
    report = "Initializing Analysis..."

    try:
        # 2. Fetch Data from Oracle 21c
        df = pd.read_sql(f"SELECT PRICE_DATE, CLOSE_PRICE FROM MARKETDATA WHERE TICKER = '{ticker}' ORDER BY PRICE_DATE", engine)
        df.columns = [c.upper() for c in df.columns]
        
        metrics = pd.read_sql(f"SELECT BETA, SHARPE_RATIO FROM RISKMETRICS WHERE TICKER = '{ticker}'", engine)
        metrics.columns = [c.upper() for c in metrics.columns]

        # 3. Pull Values (CFA: Systematic Risk & Risk-Adjusted Return)
        beta = metrics['BETA'].iloc[0] if not metrics.empty else 0.0
        sharpe = metrics['SHARPE_RATIO'].iloc[0] if not metrics.empty else 0.0

        # 4. Generate the Predictive Trend Plot
        df['SMA20'] = df['CLOSE_PRICE'].rolling(window=20).mean()
        ax.plot(df['PRICE_DATE'], df['CLOSE_PRICE'], label='Market Price', color='#004080', alpha=0.7)
        ax.plot(df['PRICE_DATE'], df['SMA20'], label='Predictive Trend (20d)', color='#ff8000', linewidth=2)
        ax.set_title(f"GS 2027 Analysis: {ticker}")
        ax.legend()

        # 5. Final CFA Risk Report
        report = f"""
        --- GS PROJECT RISK REPORT ---
        Ticker: {ticker}
        Beta (Systematic Risk): {beta:.2f}
        Sharpe Ratio: {sharpe:.2f}
        Status: {'Outperforming' if sharpe > 1 else 'Market Neutral'}
        """
        
    except Exception as e:
        report = f"Error during analysis: {e}"
    
    return fig_price, report

# Launch the Master Interface
demo = gr.Interface(
    fn=master_analytics,
    inputs=gr.Dropdown(choices=list(nifty50_list), label="Select NIFTY Asset"),
    outputs=[gr.Plot(label="Market Trends"), gr.Textbox(label="CFA Risk Metrics")],
    title="GS 2027: Enterprise Predictive Dashboard",
    description="Live Oracle 21c Backend | Predictive Analytics Engine | CFA Risk Suite"
)
demo.launch(inline=True)
```
